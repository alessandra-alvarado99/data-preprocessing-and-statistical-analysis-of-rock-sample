```python
number_of_missing_values= df_1.isnull().sum()
print(number_of_missing_values)
```

```python
df_1= df_1.dropna()
df_1.head()
```

```python
print(df_1.shape)
```

```python
sample_depth=df_1.iloc[: ,0]
insitu_permeability = df_1.iloc[:,1]
routine_porosity= df_1.iloc[:,2]
```

```python
print(insitu_permeability)
```

```python
df_2= pd.concat([sample_depth, insitu_permeability, routine_porosity], axis=1)
df_2.columns= ['Sample depth (ft)', 'In-situ permeability (mD)', 'Routine porosity (%)']
df_2.head()
```

```python
plt.boxplot(df_2['In-situ permeability (mD)'])
plt.xlabel('In-situ Klinkenberg permeability (mD)')
plt.show()
```

```python
plt.hist(df_2['In-situ permeability (mD)'], bins=3)
plt.xlabel('In-situ Klinkenberg permeability (mD)')
plt.ylabel('Frequency')
plt.show()
```

```python
Q1 = df_2['In-situ permeability (mD)'].quantile(0.25)
Q3 = df_2['In-situ permeability (mD)'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df_2_cleaned = df_2[(df_2['In-situ permeability (mD)'] >= lower_bound) & (df_2['In-situ permeability (mD)'] <= upper_bound)]
plt.hist(df_2_cleaned['In-situ permeability (mD)'], bins=3)
plt.xlabel('In-situ Klinkenberg permeability (mD)')
plt.ylabel('Frequency')
plt.show()
```

```python
df_2_cleaned.head()
```

```python
plt.boxplot(df_2_cleaned['In-situ permeability (mD)'])
plt.xlabel('In-situ Klinkenberg permeability (mD)')
plt.show()
```

```python
num_outiliers= len(df_2) - len(df_2_cleaned)
print(num_outiliers)
```

# Feature-engineering methods

```python
#Decompose features
df_2_cleaned= df_2_cleaned.copy()
df_2_cleaned['Sample depth (ft)']= df_2_cleaned['Sample depth (ft)'].str.extract('-\s*(\d+\.?\d*)\s*ft').astype(float)
df_2_cleaned.head()
```

```python
sns.pairplot(df_2_cleaned)
plt.show()
```

```python
df_2_cleaned['In-situ permeability (mD)'] = pd.to_numeric(
    df_2_cleaned['In-situ permeability (mD)'], errors='coerce'
)
df_2_cleaned.head()
```

```python
df_2_cleaned['In-situ permeability (mD)']= np.log10(df_2_cleaned['In-situ permeability (mD)'])
df_2_cleaned.head()
```

```python
sns.pairplot(df_2_cleaned)
plt.show()
```

```python
df_2_cleaned.corr()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df_2_cleaned_scaled = df_2_cleaned.copy()
df_2_cleaned_scaled[['Sample depth (ft)', 'In-situ permeability (mD)', 'Routine porosity (%)']] = scaler.fit_transform(df_2_cleaned_scaled[['Sample depth (ft)', 'In-situ permeability (mD)', 'Routine porosity (%)']])
```
